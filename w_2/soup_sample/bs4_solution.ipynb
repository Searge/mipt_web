```python
from bs4 import BeautifulSoup
import re
import os
```

```python
start = 'Stone_Age'
end = 'Python_(programming_language)'
path = './wiki/'
link_re = re.compile(r"(?<=/wiki/)[\w()]+")
```

```python
os.listdir(path)
```

```python
files = dict.fromkeys(os.listdir(path), False); files
```

```python
current_links = [start]
while current_links:
        new_links = list()
        for name in current_links:
            with open(f'{path}{name}') as data:
                links = re.findall(link_re, data.read())
            for link in links:
                if files.get(link) is False:
                    files[link] = name
                    if link == end:
                        print(files)
                    new_links.append(link)
        current_links = new_links
```
