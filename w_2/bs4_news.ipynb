```python
from bs4 import BeautifulSoup
import requests
```

```python
result = requests.get('https://ukr.segodnya.ua/')
html = result.text
soup = BeautifulSoup(html, 'lxml')
```

```python
soup.find_all('h2')
```

```python
[
    section.string for section in soup.find_all('h2')
]
```

```python
[
    (section.string,
    [
        link.string for link in section.find_parents()[3].find_all('h3') 
        if link.string is not None
    ])
    for section in soup.find_all('h2')
]
```
